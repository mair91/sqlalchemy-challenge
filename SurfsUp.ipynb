```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
import datetime as dt
```

Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Meas = Base.classes.measurement
Stat = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from the last data point in the database
lastDate = (session.query(Meas.date)
                     .order_by(Meas.date.desc())
                     .first())
lastDateNum = dt.datetime.strptime(list(np.ravel(lastDate))[0], '%Y-%m-%d')

lastY = int(dt.datetime.strftime(lastDateNum, '%Y'))
lastM = int(dt.datetime.strftime(lastDateNum, '%m'))
lastD = int(dt.datetime.strftime(lastDateNum, '%d'))

yearAgo = dt.date(lastY, lastM, lastD) - dt.timedelta(days=365)


# Perform a query to retrieve the data and precipitation scores
precData = (session.query(Meas.date, Meas.prcp)
                  .order_by(Meas.date).filter(Meas.date > yearAgo)
                  .all())

# Save the query results as a Pandas DataFrame and set the index to the date column
# Sort the dataframe by date
precDF = pd.DataFrame(precData).set_index('date').sort_index(ascending=True)

# Use Pandas Plotting with Matplotlib to plot the data
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 5.5)
plt.plot(precDF)
plt.title("Last 12 months of Precipitation")
plt.xlabel("Date")
plt.ylabel("Precipitation")
plt.xticks([])
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
precDF.describe()
```

```python
# Design a query to show how many stations are available in this dataset?
stationsNum = session.query(Stat).count()
print(f"There are {stationsNum} stations")
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.

stationsActivity = (session.query(Meas.station, func.count(Meas.station))
                        .order_by(func.count(Meas.station).desc()).group_by(Meas.station)
                        .all())
stationsActivity
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
activeID = stationsActivity[0][0]
activeName = (session.query(Stat.name).filter_by(station = activeID))
activeName = activeName[0][0]
tempHigh = (session.query(Meas.tobs).filter(Meas.station == activeID).order_by(Meas.tobs.desc()).first())
tempHigh = tempHigh[0]
tempLow = (session.query(Meas.tobs).filter(Meas.station == activeID).order_by(Meas.tobs.asc()).first())
tempLow = tempLow[0]
tempAvg = (session.query(func.avg(Meas.tobs)).filter(Meas.station == activeID))
tempAvg = '{0:.3}'.format(tempAvg[0][0])
print(f"Most Active Station: {activeID} | {activeName}")
print(f"Lowest temperature in {activeID} station: {tempLow} degrees Fahrenheit")
print(f"Highest temperature in {activeID} station: {tempHigh} degrees Fahrenheit")
print(f"Average temperature in {activeID} station: {tempAvg} degrees Fahrenheit")
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
activeStatData = (session.query(Meas.date, Meas.tobs).order_by(Meas.date)
                   .filter(Meas.date > yearAgo).filter(Meas.station == activeID)
                   .all())
activeStatDF = pd.DataFrame(activeStatData).set_index('date').sort_index(ascending=True)
activeStatDF.plot(kind='hist', bins=12)
```
